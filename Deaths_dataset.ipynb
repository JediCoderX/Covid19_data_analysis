```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
deaths_dataset_csv = pd.read_csv("Datasets/covid19_deaths_dataset.csv")
deaths_dataset_csv.head(10)
```

```python
deaths_dataset_csv.drop(["Lat", "Long"], axis=1, inplace=True)
deaths_dataset_csv.head(10)
```

```python
deaths_dataset_aggregated = deaths_dataset_csv.groupby("Country/Region").sum()
deaths_dataset_aggregated.head(10)
```

```python
deaths_dataset_aggregated.shape
```

```python
deaths_dataset_aggregated.loc["Vietnam"].plot()
plt.legend()
```

```python
deaths_dataset_aggregated.loc["Vietnam"].plot()
deaths_dataset_aggregated.loc["China"].plot()
deaths_dataset_aggregated.loc["US"].plot()
plt.legend()
```

```python
deaths_dataset_aggregated.loc["Vietnam"].diff().plot()
deaths_dataset_aggregated.loc["China"].diff().plot()
deaths_dataset_aggregated.loc["US"].diff().plot()
plt.legend()
```
